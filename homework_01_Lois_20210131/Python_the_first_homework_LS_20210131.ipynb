```python
#import pandas as pd
from pandas import Series,DataFrame
data = {'Chinese': [66, 95, 93, 90,80], 'Math': [65, 76, 86, 88, 90], 'English': [30, 98, 88, 77, 90]}
df1 = DataFrame(data)
df2 = DataFrame(data, index=['ZhangFei', 'GuanYu', 'LiuBei', 'DianWei', 'XuChu'], columns=['Chinese', 'Math', 'English'])
print(df1)
print(df2)
#print(df2.describe())
print(df2.mean(),df2.min(),df2.max(),df2.std(),df2.var())
#将每列求和并添加到表尾
#df2.loc['Row_sum'] = df2.apply(lambda x: x.sum())
#将每行求和，计算总成绩
df2['Col_sum'] = df2.apply(lambda x: x.sum(), axis=1)
print(df2)
df2.sort_values('Col_sum', ascending=False)
```
